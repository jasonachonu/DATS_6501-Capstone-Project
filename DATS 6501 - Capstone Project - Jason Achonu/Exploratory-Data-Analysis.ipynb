```python
weekday_group = usage_df.groupby(by=['date','time']).mean()
weekday_group.reset_index(inplace=True)
```

```python
def return_weekday(d_date):

    import datetime

    weekno = d_date.weekday()

    if weekno<5:
        return "Weekday"
    else:
        return "Weekend"
```

```python
weekday_group['weekday'] = weekday_group['date'].map(return_weekday)
```

```python
weekday_group.shape
```

```python
weekday_group.head()
```

```python
is_weekday = weekday_group[weekday_group['weekday'] == 'Weekday']
```

```python
not_weekday = weekday_group[weekday_group['weekday'] != 'Weekday']
```

```python
is_weekday = is_weekday.groupby(by =['hour']).mean().mean(axis = 1)
```

```python
not_weekday = not_weekday.groupby(by =['hour']).mean().mean(axis = 1)
```

```python
not_weekday = not_weekday.to_frame()
is_weekday = is_weekday.to_frame()

is_weekday.reset_index(inplace=True)
not_weekday.reset_index(inplace=True)

is_weekday.columns = ['hour','daily mean']
not_weekday.columns = ['hour','daily mean']
```

```python
not_weekday['weekday'] = 'Weekend'
is_weekday['weekday'] = 'Weekday'
```

```python
all_week = pd.concat([is_weekday,not_weekday])
```
